## Benchmark on several graphs

```typescript
let N_nodes = d3.range(10, 100, 10)
```

```typescript
for (let n_nodes of N_nodes) {
    console.log(n_nodes)
    let fp = `../data/SBM/${n_nodes}Nodes_${n_nodes}edges_weighted.json`;
    const text = await Deno.readTextFile(fp);
    const data = JSON.parse(text);

    let graph = Hypergraph.fromJson(data)
    let layout = new HypForceLayout(graph, WIDTH, HEIGHT);
    layout.run();

    const overlapArea = computeHypergraphOverlapArea(graph);
    console.log("OVERLAP AREA", overlapArea);
    const area = computeHypergraphArea(graph);
    console.log("AREA", area);
    console.log("NORMALIZED OVERLAP AREA", overlapArea / area);
}
```
