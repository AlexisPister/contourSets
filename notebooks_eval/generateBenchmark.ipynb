## Generate Benchmark

### Random Hypergraph

```typescript
let Ns = [];
let p = 0.1;

for (let i = 10; i <= 1000; ) {
  Ns.push(i);

  if (i < 100) {
    i += 10;   // 10, 20, ..., 100
  } else {
    i += 100;  // 200, 300, ..., 1000
  }
}
```

```typescript
Ns
```

```typescript
for (let n of Ns) {
    let graph = randomHypergraph(n, n, p);
    let graphJson = graph.toJson();

    Deno.writeTextFile(`random${n}_${p}.json`, JSON.stringify(graphJson, null, 2));
}
```

### Preferential Attachment

```typescript
for (let n of Ns) {
    let graph = prefattachmentHypergraph(n, n, p);
    let graphJson = graph.toJson();

    Deno.writeTextFile(`pref${n}_${p}.json`, JSON.stringify(graphJson, null, 2));
}
```
